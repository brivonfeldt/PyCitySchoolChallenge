```python
## import pandas as pd
import numpy as np
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "./schools_complete.csv"
student_data_to_load = "./student_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
#student_data.count() #=39170
    
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_info_dicts = ({school_data_complete})
school_data_complete_df = pd.DataFrame(school_data_complete)
del(school_data_complete_df['School ID'])
school_data_complete_df.head()
#school_data_complete_df.count()
```

```python
student_count_df = pd.DataFrame(school_data_complete_df.groupby(['school_name', 'type', 'budget'])['student ID'].count()).reset_index()
student_count_df.head()
```

```python
score_avrg_df = pd.DataFrame(school_data_complete_df.groupby(['school_name'])['math_score', 'reading_score'].mean()).reset_index()
score_avrg_df.head()
```

```python
#FINDING THE TOTAL NUMBER OF SCHOOLS 
unique_schools_list = school_data_complete_df['school_name'].unique()
total_count_of_unique_schools = len(unique_schools_list)
#print(total_count_of_unique_schools)
```

```python
#FINDING THE TOTAL NUMBER OF STUDENTS
student_count = len(school_data_complete_df['student ID'])
student_count
#count = school_data_complete_df['school_name'].count()
# print(count)
```

```python
#FINDING THE SUM OF THE TOTAL SCHOOLS BUDGET
total_county_wide_school_budget = school_data_complete_df['budget'].unique().sum()
print(f'The total budget of all of the schools are : ${total_county_wide_school_budget}.00')
```

```python
#FINDING THE AVERAGE MATH SCORES
avg_math_score_total_student_names = school_data_complete_df['math_score'].mean()
avg_math_score_total_student_names
```

```python
#FINDING THE AVERAGE READING SCORES
avg_reading_score_total_student_names = school_data_complete_df['reading_score'].mean()
avg_reading_score_total_student_names 
```

```python
# PERCENTAGE OF STUDENTS WITH PASSING MATH SCORES
math_score_list = school_data_complete_df['math_score']
#shows this index in the series as and int64 
passing_math_scores = []
#type(math_score_list[0])
for row in math_score_list:
    if row > 70:
        passing_math_scores.append(row)
passing_math_scores_count = len(passing_math_scores)
# passing_math_scores_count
passing_math_avg = (passing_math_scores_count/student_count)*100
passing_math_avg
```

Correct the output formatting above ^^^^^^^^^

```python
# PERCENTAGE OF STUDENTS WITH PASSING READING SCORES
reading_score_list = school_data_complete_df['reading_score']
#shows this index in the series as and int64 
passing_reading_scores = []
#type(math_score_list[0])
for row in reading_score_list:
    if row >= 70:
        passing_reading_scores.append(row)
passing_reading_scores_count = len(passing_reading_scores)
#passing_reading_scores_count
passing_reading_avg = (passing_reading_scores_count/student_count)*100
passing_reading_avg
```

```python
#OVERALL AVERAGE
overall_avg = (passing_reading_avg+passing_math_avg)/2
overall_avg
```

```python
passing_math_scores_df = school_data_complete_df.loc[school_data_complete_df['math_score'] >= 70]
# passing_math_scores_df.head()
passing_math_scores_groupby_df = pd.DataFrame(passing_math_scores_df.groupby(['school_name'])['math_score'].count().reset_index())
passing_math_scores_groupby_df.head()
```

```python
passing_reading_scores_df = school_data_complete_df.loc[school_data_complete_df['reading_score'] >= 70]
# passing_reading_scores_df.head()
passing_reading_scores_groupby_df = pd.DataFrame(passing_reading_scores_df.groupby(['school_name'])['reading_score'].count().reset_index())
passing_reading_scores_groupby_df.head()
```

```python
#Create a new df with schools as the index
school_summary_df = pd.DataFrame({
    'School Name' : student_count_df['school_name'], 
    'School Type' : student_count_df['type'],
    'Total Students' : student_count_df['student ID'],
    'Total School Budget' : student_count_df['budget'],
    'Per Student Budget' : student_count_df['budget'] / student_count_df['student ID'], 
    'Avg Math Score' : score_avrg_df['math_score'],
    'Avg Reading Score' : score_avrg_df['reading_score'],
    'Passing Math' : passing_math_scores_groupby_df['math_score'] / student_count_df['student ID']*100,
    'Passing Reading' : passing_reading_scores_groupby_df['reading_score'] / student_count_df['student ID']*100,
    'Overall Passing' : (((passing_math_scores_groupby_df['math_score'] / student_count_df['student ID'])
                        + (passing_reading_scores_groupby_df['reading_score'] / student_count_df['student ID'])) / 2)*100 
})

school_summary_df
```

```python
top_schools_df = school_summary_df.sort_values(by='Overall Passing', ascending=False)
top_schools_df
```

```python
bottom_schools_df = school_summary_df.sort_values(by='Overall Passing', ascending=True)
bottom_schools_df
```
